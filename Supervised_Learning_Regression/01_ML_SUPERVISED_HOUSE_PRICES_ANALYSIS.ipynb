## Abstract
Housing price trends within a district constitute critical information for policymakers and financial institutions. Unfortunately, existing house price estimates from the State of California exhibit significant inaccuracies, exceeding 20%, due to reliance on complex rules and outdated models. Consequently, enhancing the accuracy of housing price estimates is paramount for effective policymaking and real estate operations. This research aims to reduce uncertainty in housing prices by developing a machine learning framework that improves estimation accuracy. To achieve this, census data from California districts were utilised and processed to identify the most suitable machine learning algorithm based on mean squared error. Results indicate that a fine-tuned Random Forest model reduced prediction errors by 50%, leading to a more than 20% improvement in estimation accuracy. Despite these promising findings, the model exhibited overfitting, as evidenced by discrepancies between training and test results. Future research will focus on mitigating this issue to produce more accurate and robust housing price estimates.

```python
#@title DATA COLLECTION (TABLES 1,2 and 3)
import os              # module for file and directory operations
import tarfile         # module to work with tar compressed files
import urllib.request  # module for downloading files from URLs

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
# Create a path
"""
It improves organisation, reproducibility, and flexibility of your code.
"""
HOUSING_PATH = os.path.join("datasets", "housing")

# Function to decompress the data
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    """
    This function , it creates a datasets/housing
    directory in your workspace, downloads the housing.tgz file, and extracts the
    housing.csv file from it in this directory.
    This function that downloads the data is useful in particular if the data changes regularly:
    you can write a small script that uses the function to fetch the latest data
    """
    # creates all necessary parent directories in the path if they don't exist.
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

# fetch_housing_data()
def load_housing_data(housing_path=HOUSING_PATH, filename="housing.csv"):
    """
    This function returns a pandas DataFrame object containing all the data.
    inputs:
    housing_path=HOUSING_PATH,
    filename="housing.csv"  # the file to load located at the specified path (housing_path)
    """
    csv_path = os.path.join(housing_path, filename)
    return pd.read_csv(csv_path)

# Get the data
# decompress data
fetch_housing_data()
# Get dataframe
housing_path=HOUSING_PATH
filename="housing.csv"
housing = load_housing_data(housing_path, filename) # Dataset
# Check data characteristics
print("\nTABLE 1: Data features Information")
housing.info()

# Check categorical data characteristics
print("\nTABLE 2: Categorical features Information")
print(housing["ocean_proximity"].value_counts())

# Check numeric data characteristics
print("\nTABLE 3: Numeric features Information")
housing.describe()
```

## 3.1.3 Data split
As it was mentioned in the methods section, the best way to select the testing dataset it is  using stratify sampling along the 80-20% rule, which means 80% is the training set, and 20% for the test set.
Following the above principle an importance analysis was conducted to assess the weight of features by examining their linear correlations with the target variable and the distribution of highly correlated feature before carrying out stratify sampling. After that, the training and testing samples were determined using the 80-20% rule. The results of the process can be observed in Table 4 and Figure 2.
Table 4 reveals that "medium_income" exhibits the highest correlation, followed by "total_rooms" to a lesser extent. Consequently, the distribution of the income feature (Figure 2) was analysed to identify categories carrying the most weight, which appear to be categories 2 and 3. These categories were incorporated into the sampling scheme to ensure a realistic distribution of input variables in the training set.
Finally, Table 5 presents the differences between the two sampling techniques. It compares the income category proportions in the overall dataset, the test set generated with stratified sampling, and the test set generated using purely random sampling. As observed, the test set generated using stratified sampling exhibits income category proportions nearly identical to those in the full dataset. In contrast, the test set generated using purely random sampling demonstrates an error of up to 4%.


```python
#@title Train and Test dataset setup (TABLE 4 & FIGURE 2).
# import function
from sklearn.model_selection import StratifiedShuffleSplit # splitting function that splits the dataset using stratified sampling
from sklearn.model_selection import train_test_split       # simple splitting function.

# Ensure groups of interest are represented(strata).
"""
Groups of interest are subsamples of a feature that has got a high impact on the target value.
Features that have significant impact on the target are identified by correlations or by prior knowledge.
In this example, it is the median income.
"""
# Simple correlation analysis for cheking influential features.
non_numeric_columns = housing.select_dtypes(include=['object', 'category']).columns # Identify non-numeric columns
numeric_housing = housing.drop(columns=non_numeric_columns)                         # Exclude non-numeric columns
corr_matrix = numeric_housing.corr()                                                # Calculate the correlation matrix
# Display the correlation with the target variable
print("\nTABLE 4: Target value correlations")
print(corr_matrix["median_house_value"].sort_values(ascending=False))

# stratified sampling to minimise bias_________________
# i. Create a income caterogy column ranging from 1.5 to 6 splitted in 5 categories(1 to 5)
print("\nFIGURE 2: Income category histogram")
housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels=[1, 2, 3, 4, 5])
housing["income_cat"].hist();#plt.title("Income category histogram")
plt.show()
```

```python
#@title Dataset spliting (TABLE 5)
# stratified sampling to minimise bias_________________

# i. Conduct stratify sampling split based on previos categories
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing["income_cat"]): # loop over dataset based on new category
    strat_train_set = housing.loc[train_index]
    strat_test_set  = housing.loc[test_index]

# ii.Non-stratified sample Split the dataset
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

# spliting techniques analysis.___________
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
print("\nTABLE 5: Stratified vs Random dataset splitting based on  income category proportions Error")
compare_props.head()
```

```python
#@title Section setup
# Import additional libraries
import seaborn as sns
from pandas.plotting import scatter_matrix

# Define functions________________________________
# Cleaning functions---------------------------------------------------
# Define function to find missing values and perform additional tasks
def find_missing_values(df):
    """
    This function identifies and prints missing values in a dataframe,
    and returns a tuple containing the location of missing values.

    Args:
        df (pd.DataFrame): The dataframe containing the data.

    Returns:
        pd.DataFrame: A dataframe containing only the rows with missing values.
        list: A list of tuples containing the location (row, column) of missing values.
    """

    # Identify missing values
    missing_values = df.isnull()

    # Total number of missing values and their percentage in the dataset
    total_missing = missing_values.sum().sum()
    total_cells = df.size
    missing_percentage_total = (total_missing / total_cells) * 100

    print(f"Total number of missing values: {total_missing}")
    print(f"Percentage of missing values in total dataset: {missing_percentage_total:.2f}%\n")

    # Calculate missing values per column and their percentage
    missing_values_per_column = missing_values.sum()
    missing_percentage_per_column = (missing_values_per_column / len(df)) * 100
    missing_percentage_total_per_column = (missing_values_per_column / total_cells) * 100

    # Create dataframe with missing values per column and their percentages
    missing_table = pd.DataFrame({
        'Missing Values': missing_values_per_column,
        'Missing Values (Per Column) %': missing_percentage_per_column,
        'Missing Values (Total Dataset) %': missing_percentage_total_per_column
    })

    # Format percentages
    missing_table['Missing Values (Per Column) %'] = missing_table['Missing Values (Per Column) %'].apply(lambda x: f'{x:.1f}')
    missing_table['Missing Values (Total Dataset) %'] = missing_table['Missing Values (Total Dataset) %'].apply(lambda x: f'{x:.1f}')

    print("Missing Values per Column:")
    print(missing_table)

    # Find the location of missing values
    missing_locations = []
    for column in df.columns:
        missing_indices = df.index[df[column].isnull()].tolist()
        missing_locations.extend([(idx, column) for idx in missing_indices])

    return df[missing_values.any(axis=1)], missing_locations

# zero value analysis
# Function to find rows with zeros in the specified columns
def find_zeros(df, col_names):
    """
    This function identifies and prints rows with zeros in the specified columns of a dataframe.
    Args:
        df (pd.DataFrame): The dataframe containing the data.
        col_names (list): List of column names to check for zeros.
    Returns:
        pd.DataFrame: A dataframe containing only the rows with zeros in the specified columns.
    """

    # Identify rows with zeros in the specified columns
    zeros_indices = df[(df[col_names] == 0).any(axis=1)].index

    # Return the rows with zeros in the specified columns
    return df.loc[zeros_indices]

# Visualisation functions ------------------------------------
# 3.22 Heatmap
def correlation_heatmap(df):
    """
    Description:
    Creates a heatmap of the correlation matrix of a DataFrame.
    Args:
        df: A DataFrame with numeric columns.
    Returns:
        A heatmap of the correlation matrix.
    """
    # Remove non-numeric columns for correlation
    df_numeric = df.select_dtypes(['float64', 'int64'])

    # Calculate the correlation matrix
    correlation_matrix = df_numeric.corr()

    # Create a heatmap of the correlation matrix
    #'ticks': np.arange(0, 1, 0.1)
    fig, ax = plt.subplots(figsize=(14, 12))
    sns.heatmap(
        correlation_matrix,
        annot=True,
        cmap='coolwarm',
        center=0,
        cbar_kws={'aspect': 40,
                  'shrink': 0.99,
                   'ticks': np.arange(-1, 1, 0.1)},  # Adjust the aspect ratio and increments of the colour bar
        ax=ax
    )

    # Adjust the rotation and position of the x-axis ticks
    ax.set_xticklabels(ax.get_xticklabels(), rotation=35, ha='right')  # Rotate x-labels to 35 degrees
    ax.set_yticklabels(ax.get_yticklabels(), rotation=0)  # Keep y-labels horizontal
    # Adjust the spacing for the title and the plot
    fig.subplots_adjust(top=0.85, bottom=0.2)
    # Set a title for the heatmap
    ax.set_title('Correlation Heatmap', fontsize=16, y=1.05)

    return fig
# Scatter plots
# Define a function to plot the scatter matrix with customized histograms
def plot_scatter_matrix(df, attributes, bins=10, figsize=(12, 8)):
    scatter_matrix_figure = scatter_matrix(df[attributes], figsize=figsize, diagonal='hist')

    # Modify histograms with the specified number of bins and edge lines
    #for i, ax in enumerate(scatter_matrix_figure.diagonal()):
    #    ax.clear()
    #    ax.hist(df[attributes[i]], bins=bins, edgecolor='black')
    #    #ax.set_title(attributes[i])

    plt.show()

#-------------------------------------------------------------------------------
# Create a copy of training set (to explore data without modify original test)
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
housing = strat_train_set.copy()
```

## 3.3 Feature analysis.
This section presents and discusses the results of the feature analysis, which aimed to identify promising features and potential combinations. Figure 8 indicates high correlations (average > 0.85) among features such as total number of rooms, bedrooms, population, and households. To address this multicollinearity, three new variables—rooms per household, bedrooms per room, and population per household—were computed as ratios of the original features. The correlation of these new features with the target variable is depicted in Figure 8.

Two of the derived features, rooms per household and bedrooms per room, exhibited stronger correlations with the median house value than their constituent features. This suggests that houses with lower bedroom-to-room ratios and larger houses (as indicated by rooms per household) tend to be more expensive. Consequently, these features were retained for subsequent analysis. In contrast, population per household demonstrated a weaker correlation (0.022) compared to its constituent features (households: 0.065, population: -0.027). This feature was therefore excluded from further consideration. These findings highlight the iterative nature of feature engineering.

```python
#@title New features (Figure 8)
housing["rooms_per_household"] = housing["total_rooms"]/housing["households"]
housing["bedrooms_per_room"] = housing["total_bedrooms"]/housing["total_rooms"]
housing["population_per_household"]=housing["population"]/housing["households"]

# Get the list of columns
columns = list(housing.columns)
# Find the index of the "median_house_value" column
median_house_value_index = columns.index("median_house_value")
# Define the new columns you want to add
new_columns = ["rooms_per_household", "bedrooms_per_room", "population_per_household"]
# Remove the new columns from their current position
for col in new_columns:
    columns.remove(col)
# Insert the new columns before the "median_house_value" column
for col in reversed(new_columns):
    columns.insert(median_house_value_index, col)
# Reorder the DataFrame
housing = housing[columns]

# Correlation matrix plot
print("\nFIGURE 8:Correlation Heatmap with new features")
fig = correlation_heatmap(housing)
plt.show()
```

To convert categorical data into a numerical format, encoding techniques were applied. The outcomes of encoding the 'ocean proximity' feature are illustrated in Table 8. As shown, ordinal encoders produce integer values, while one-hot encoders generate binary representations. The rationale behind this is that ordinal encoders are typically utilized when an ordinal relationship exists among categories, whereas one-hot encoders are employed when no clear order is present. Given the presence of only one categorical feature in this dataset, resulting in minimal dimensionality concerns, one-hot encoding was selected for this analysis.

The final stage of data preprocessing is summarized in Table 9, which presents the transformed dataset ready for input into a machine learning model. This dataset incorporates both standardised numerical features and one-hot encoded categorical features within a unified structure.

```python
#@title Custom Transformers (Table 9)
# Create a custom transformer to add extra attributes:
from sklearn.base import BaseEstimator, TransformerMixin

# get the right column indices: safer than hard-coding indices 3, 4, 5, 6
rooms_ix, bedrooms_ix, population_ix, household_ix = [
    list(housing.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", "population", "households")]

# Define custom transformer
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kwargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
# Transform data
housing_extra_attribs = attr_adder.transform(housing.values)
# Transform np array to pd
housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(housing.columns)+["rooms_per_household", "population_per_household"],
    index=housing.index)

print("\nTABLE 9: Dataframe New features")
housing_extra_attribs.head()
```

```python
#@title Final pipeline (Table 10)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer  # apply different preprocessing steps to different subsets of features.

# Numeric pipeline (only numeric transformations)
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),                     # scaling (standardise features) numerical features
])

housing_num_tr = num_pipeline.fit_transform(housing_num)

# Mix pipeline (handles numeric and categorical transformations together)
num_attribs = list(housing_num)   # numerical attributes
cat_attribs = ["ocean_proximity"] # categorical attributes
full_pipeline = ColumnTransformer([
                                    ("num", num_pipeline, num_attribs),
                                    ("cat", OneHotEncoder(), cat_attribs),
                                  ])

housing_prepared = full_pipeline.fit_transform(housing) # Data to be input to model
print("\nTABLE 10: Final array to be inputed in Models")
housing_prepared # Array
```

## 3.6 Finetuning
After selecting the best model, the next step is to analyse the best combination of hyper-parameters for the previous models. Scikit-Learn’s GridSearchCV and RandomizedSearchCV are functions that automatises such task. Whilst, the grid search approach is fine when you are exploring relatively few combinations, when the hyperparameter search space is large, it is often preferable to use RandomizedSearchCV instead. The results can be seen in Table 16, in which the max combination of features are 8 and 7, and the number of estimators, i.e. tree splits, are 30 and 180 for the grid and random searches respectively.  Whilst more trees bring more accuracy, there is a higher computational cost; consequently, the best combination is normally given by minimum RMSE value. Given the reduced feature set in this study, grid search, which employs eight features and thirty tree splits , was selected as computational performance was not a significant constraint for  better model accuracy.


```python
#@title setupt
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

```

```python
#@title Hyperparameters tuning (Table 16)

# Grid search
param_grid = [
    # try 12 (3×4) combinations of hyperparameters
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # then try 6 (2×3) combinations with bootstrap set as False
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

forest_reg = RandomForestRegressor(random_state=42)
# train across 5 folds, that's a total of (12+6)*5=90 rounds of training
grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error', return_train_score=True)
grid_search.fit(housing_prepared, housing_labels)

# Best combination
#print("\nThe best combination of Hyperparameters for the RF is:")
#print(grid_search.best_params_)

# Random model_______________________________________________________
param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(housing_prepared, housing_labels)
# Results--------------------------------------------------------------
print("\nTABLE 16: Combination of Hyperparameters:\n")
# Best combination
print("\nGrid search best combination of Hyperparameters for the RF is:")
print(grid_search.best_params_)
# Best combination random model (random model)
print("\nRandom search best combination of Hyperparameters is:")
print(rnd_search.best_params_)
```

## 3.9 Validation
This section evaluates the accuracy of the proposed model using a completely independent test dataset. Tables 18 and 19 present the results. The MRG model exhibited minimal variation between the training and test datasets, indicating a high level of model accuracy. In contrast, the final ML model displayed a one-fold increase in error. This increase is attributable to overfitting, a common issue in tree-based models. Potential mitigations for overfitting include further hyperparameter tuning, such as limiting tree depth or reducing the number of trees. Additionally, collecting more data and reducing the feature set could help to narrow the gap between training and testing RMSE. However, given the significant performance improvement of the proposed ML model over the MRG model, further mitigation strategies were not explored in this study.

```python
#@title  Validation (Table 18 and 19)
# ML MODEL
# Set testing set
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()
# Predictions
X_test_prepared = full_pipeline.transform(X_test)
final_predictions = final_model.predict(X_test_prepared)
# Metrics
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

# Confidence intervals
confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
mean = squared_errors.mean()
m = len(squared_errors)
# results
print("TABLE 18: Final RF testing results");
print("\nFinal RF RMSE");print(final_rmse)
print("\nFinal RF RMSE Confidence interval")
print(np.sqrt(stats.t.interval(confidence, m - 1,
                         loc=np.mean(squared_errors),
                         scale=stats.sem(squared_errors))))
#-------------------------------------------------------------
# INDUSTRY MODEL TEST
# Set testing set
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()
#
df_test = pd.DataFrame(strat_test_set[['total_rooms','median_income','median_house_value']].copy())
# Quadratic terms
df_test['total_rooms_squared'] = df_test['total_rooms']**2
df_test['median_income_squared'] = df_test['median_income']**2

# Define the independent variables (including quadratic terms)
XR_TEST = df_test[['total_rooms', 'total_rooms_squared', 'median_income', 'median_income_squared']]
XR_TEST = sm.add_constant(XR_TEST)  # Add a constant term for the intercept
# Make predictions on the test set
predictions_XR_TEST = current_model.predict(XR_TEST)
# CI
# Confidence intervals
confidence = 0.95
squared_errors = (predictions_XR_TEST - y_test) ** 2
mean = squared_errors.mean()
m = len(squared_errors)

# Metrics RG
print("\nTABLE 19: Final MRG testing results");
rtest_mse = mean_squared_error(y_test, predictions_XR_TEST)
rtest_rmse = np.sqrt(rtest_mse)
print("\nFinal MRG RMSE (test set):\n", rtest_rmse)
print("\nFinal MRG Confidence interval")
np.sqrt(stats.t.interval(confidence, m - 1,
                         loc=np.mean(squared_errors),
                         scale=stats.sem(squared_errors)))
```
