The different metrics are described below.

### 2.5.1 Confusion Matrix (table)
It is a table with two rows and two columns, for binary classifiers, that reports the number of true positives(TP), false negatives(FN), false positives(FP), and true negatives(TN). In binary classifiers, the classifier tries to predict the actual class, in which 1 is the positive, i.e. belongs to the class, and 0 is the negative, i.e. does not belong to the class.

The confusion matrix metric allows more detailed analysis than simply observing the proportion of correct classifications (accuracy). Accuracy will yield misleading results if the data set is unbalanced; that is, when the numbers of observations in different classes vary greatly.

### 2.5.2 Positive Class Metrics
#### Precision
Precision is the ratio of correctly predicted positive observations to the total predicted positives. It measures the accuracy of the positive predictions.
$$Precision = \frac{TP}{TP + FP}$$

#### Recall(Sensitivity)
Also known as true positive rate (TTR), and sensitivity, it is the ratio of correctly predicted positive observations to all the actual positives.
$$Recall = \frac{TP}{TP + FN}$$

#### F-1 Score
The F1 score is the harmonic mean of precision and recall. It balances the trade-off between precision and recall.
$$F1 = 2 * \frac{precision * recall}{precision + recall}$$

### 2.5.3 Negative class metrics
#### Negative Predictive Value (NPV)
NPV measures the accuracy of negative predictions.
$$NPV = \frac{TN}{TN + FN}$$

#### Specificity
It is the proportion of actual negatives that are correctly identified as such.
$$Specificity = \frac{TN}{TN + FP}$$

### ROC and AUC
#### ROC (Receiver Operating Characteristic)
curve is a plot of the true positive rate (TPR) against the false positive rate (FPR) for various classification thresholds.  
$$FPR = \frac{FP}{FP + TN}$$

#### AUC (Area Under the Curve)
It is the area under the ROC curve. It provides an aggregate measure of performance across all classification thresholds.
Interpretation: Higher AUC indicates better performance.
Note: ROC curves and AUC are particularly useful for imbalanced datasets.

```python
#@title setup
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
from sklearn.base import BaseEstimator
from sklearn.dummy import DummyClassifier
# Cofusion matrix
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
# Precision and recall
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve
# ROC
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
# others
import matplotlib.patches as patches  # extra code – for the curved arrow
# Functions
def plot_precision_recall_vs_threshold(precisions, recalls, thresholds, threshold):
    """
    This function plots the precision-recall curve for a given classification task.
    Args:
        precisions (list): List of precision values for each threshold.
        recalls (list): List of recall values for each threshold.
        thresholds (list): List of thresholds used to calculate precision and recall.
        threshold (float): Specific threshold value to highlight on the plot.
    """
    plt.figure(figsize=(8, 4))  # extra code – it's not needed, just formatting

    # Find the intersection point (where precision equals recall)
    idx_intersection = np.argwhere(np.isclose(precisions[:-1], recalls[:-1], atol=0.01)).flatten()
    intersection_threshold = thresholds[idx_intersection][0] if len(idx_intersection) > 0 else None
    intersection_precision = precisions[idx_intersection][0] if len(idx_intersection) > 0 else None
    intersection_recall = recalls[idx_intersection][0] if len(idx_intersection) > 0 else None

    # Highlight the specific threshold point
    idx = (thresholds >= threshold).argmax()  # First index greater than or equal to threshold
    precision = precisions[idx]
    recall = recalls[idx]
    plt.vlines(threshold, 0, 1.0, "k", "dotted", label=f"Threshold ({threshold})")
    plt.plot(thresholds[idx], precisions[idx], "bo", markersize=8)
    plt.plot(thresholds[idx], recalls[idx], "go", markersize=8)

    # Plot recall and precision
    plt.plot(thresholds, precisions[:-1], "b--", label=f"Precision={precision:.2f}", linewidth=2)
    plt.plot(thresholds, recalls[:-1], "g-", label=f"Recall={recall:.2f}", linewidth=2)

    # Highlight the intersection point if it exists
    if intersection_threshold is not None:
        #plt.plot(intersection_threshold, intersection_precision, "ro", markersize=8, label="Intersection Point")
        plt.plot(intersection_threshold, intersection_precision, "ro", markersize=8, label=f"P/R={intersection_precision:.2f},Threshold={intersection_threshold:.2f}")
        #plt.annotate(f"({intersection_threshold:.2f}, {intersection_precision:.2f})",
        #             (intersection_threshold, intersection_precision),
        #            textcoords="offset points", xytext=(10,-10), ha='center', color='red')

    plt.axis([min(thresholds)*0.2, max(thresholds)*0.6, 0, 1])  # Set axis limits based on thresholds
    plt.grid()
    plt.xlabel("Threshold")
    plt.ylabel("Precision/Recall")  # Combined label for both curves
    plt.legend(loc="center right")

    # You can customize this section for saving or displaying the plot (optional)
    plt.tight_layout()
    plt.show()
# Functions

```

**ROC and AUC**

The ROC (Receiver Operating Characteristic) curve and corresponding AUC (Area Under the Curve) metrics are presented in Table 16 and Figure 8. The ROC curve plots the true positive rate (TPR) against the false positive rate (FPR) for various classification thresholds. The AUC provides an aggregate measure of performance across all classification thresholds. A perfect classifier achieves an AUC of 1, while a random classifier yields an AUC of 0.5.
