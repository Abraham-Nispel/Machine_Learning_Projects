The different metrics are described below.

### 2.5.1 Confusion Matrix (table)
It is a table with two rows and two columns, for binary classifiers, that reports the number of true positives(TP), false negatives(FN), false positives(FP), and true negatives(TN). In binary classifiers, the classifier tries to predict the actual class, in which 1 is the positive, i.e. belongs to the class, and 0 is the negative, i.e. does not belong to the class.

The confusion matrix metric allows more detailed analysis than simply observing the proportion of correct classifications (accuracy). Accuracy will yield misleading results if the data set is unbalanced; that is, when the numbers of observations in different classes vary greatly.

### 2.5.2 Positive Class Metrics
#### Precision
Precision is the ratio of correctly predicted positive observations to the total predicted positives. It measures the accuracy of the positive predictions.
$$Precision = \frac{TP}{TP + FP}$$

#### Recall(Sensitivity)
Also known as true positive rate (TTR), and sensitivity, it is the ratio of correctly predicted positive observations to all the actual positives.
$$Recall = \frac{TP}{TP + FN}$$

#### F-1 Score
The F1 score is the harmonic mean of precision and recall. It balances the trade-off between precision and recall.
$$F1 = 2 * \frac{precision * recall}{precision + recall}$$

### 2.5.3 Negative class metrics
#### Negative Predictive Value (NPV)
NPV measures the accuracy of negative predictions.
$$NPV = \frac{TN}{TN + FN}$$

#### Specificity
It is the proportion of actual negatives that are correctly identified as such.
$$Specificity = \frac{TN}{TN + FP}$$

### ROC and AUC
#### ROC (Receiver Operating Characteristic)
curve is a plot of the true positive rate (TPR) against the false positive rate (FPR) for various classification thresholds.  
$$FPR = \frac{FP}{FP + TN}$$

#### AUC (Area Under the Curve)
It is the area under the ROC curve. It provides an aggregate measure of performance across all classification thresholds.
Interpretation: Higher AUC indicates better performance.
Note: ROC curves and AUC are particularly useful for imbalanced datasets.

Finally, the results for data augmentation scheme developed for this project are illustrated in Tables 12 and 13 and Figure 5.Data augmentation expands the dataset by generating new data points through techniques such as bootstrapping or, in the case of images, rotating or transforming existing data. In this study, new data was introduced by shifting the existing classes clockwise, as depicted in Figure 5. Training the model with this augmented dataset resulted in a 17% increase in accuracy compared to a model using only the original dataset.


```python
#@title setup
# finetuning
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from scipy.ndimage import shift

# Functions--------------------------
# Data augmentation
def shift_image(image, dx, dy):
    image = image.reshape((28, 28))
    shifted_image = shift(image, [dy, dx], cval=0, mode="constant")
    return shifted_image.reshape([-1])
```

```python
#@title Finetuning of Hyperparameters (TABLE 10)
print("TABLE 10: Model Hyperparameters Tuning\n")
# Set hyper-parameters (knn  are weights, and number of neighbours)
param_grid = [{'weights': ["uniform", "distance"], 'n_neighbors': [3, 4, 5, 6]}]
# Find parameters with 10.000 images (small sample for reducing comp time)
knn_clf = KNeighborsClassifier()
grid_search = GridSearchCV(knn_clf, param_grid, cv=5)
grid_search.fit(X_train_small, y_train_small)
print("The KNN best parameters are:")
print(grid_search.best_params_)

# Print new accuracy
grid_search.best_estimator_.fit(X_train_small, y_train_small)
tuned_accuracy = grid_search.score(X_test_small, y_test_small)
print("\nModel accuracy after tuning:\n",tuned_accuracy)
```

```python
#@title Data scaling (TABLE 11)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_small.astype("float64"))
X_test_scaled = scaler.transform(X_test_small.astype("float64"))

# Accuracy cross validation (takes a long time)
grid_search.best_estimator_.fit(X_train_scaled, y_train_small)
scaled_accuracy = grid_search.score(X_test_scaled, y_test_small)
print("TABLE 11: KNN Scaled data CV Accuracy")
print(scaled_accuracy)

#print("TABLE 9: KNN Scaled data CV Accuracy")
#print(cross_val_score(grid_search, X_train_scaled, y_train_small, cv=3, scoring="accuracy"))
```

```python
#@title Data augmentation (FIGURE 5)
print("FIGURE 5: Zero class shifting")
# Augmented dataset---------------------------------------
X_train_augmented = [image for image in X_train]
y_train_augmented = [label for label in y_train]

for dx, dy in ((-1, 0), (1, 0), (0, 1), (0, -1)):
    for image, label in zip(X_train, y_train):
        X_train_augmented.append(shift_image(image, dx, dy))
        y_train_augmented.append(label)

X_train_augmented = np.array(X_train_augmented)
y_train_augmented = np.array(y_train_augmented)

# Shuffle the new data with current data
shuffle_idx = np.random.permutation(len(X_train_augmented))
X_train_augmented = X_train_augmented[shuffle_idx]
y_train_augmented = y_train_augmented[shuffle_idx]

# Plot
image = X_train[1000]  # some random digit to demo
shifted_image_down = shift_image(image, 0, 5)
shifted_image_left = shift_image(image, -5, 0)

plt.figure(figsize=(12, 3))
plt.subplot(131)
plt.title("Original")
plt.imshow(image.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.subplot(132)
plt.title("Shifted down")
plt.imshow(shifted_image_down.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.subplot(133)
plt.title("Shifted left")
plt.imshow(shifted_image_left.reshape(28, 28),
           interpolation="nearest", cmap="Greys")
plt.show()
```

```python
#@title Accuracy analysis (TABLE 12 and 13)
# Original dataset accuracy
knn_final_clf = KNeighborsClassifier(**grid_search.best_params_)
print("TABLE 12: KNN Full training data CV Accuracy")
knn_final_clf.fit(X_train, y_train)
full_accuracy = knn_final_clf.score(X_test, y_test)
print(full_accuracy)

# Accuracy augmented
#knn_clf = KNeighborsClassifier(**grid_search.best_params_)
knn_final_clf.fit(X_train_augmented, y_train_augmented)
augmented_accuracy = knn_final_clf.score(X_test, y_test)
print("\nTABLE 13: KNN Augmented data CV Accuracy")
print(augmented_accuracy)
print("new dataset length",len(X_train_augmented))

# Error analysis
error_rate_change = 1- (1 - augmented_accuracy) / (1 - full_accuracy)
print(f"\nerror_rate_change = {error_rate_change:.0%}")
```

Another important element to considerate to ensure the robustness of the model, it is the analysis of normalised errors, which can be seen in Figure 7. These plots were created by dividing each value in the confusion matrix by the number of misclassified images in the corresponding class to enable comparison of error rates rather than absolute numbers of errors (which could unfairly penalise abundant classes). The diagonal was filled with zeros to isolate errors, and the rows represent actual classes, while columns represent predicted classes. From the figure, it is evident that the classifier's error patterns; for example, 4s and 9s, which is frequently confused in both directions, presented error rates of 63% and 48% respectevely. An explanation for the errors can be attributed to the poor technique presented by students that may challenge the model capabilities (see Figure 8). In addition, 1s and 2s, or 7s and 1s classes, tend to be confused in one direction- that is, 7 is predicted as 1, when in reality it is one.

To enhance the classifier's performance, efforts should be focused on reducing the confusion between 9s and 4s. This could involve collecting more data, engineering new features, or employing more sophisticated image processing techniques. However, given the time-consuming and complex nature of these approaches, they are beyond the scope of this report.

Finally, Table 12 summarises the accuracy metrics of the final model. The AUC surpasses 99%, demonstrating that the model has a strong ability to distinguish between the positive and negative classes. These results suggest that the model has a low rate of false positives (errors where the model incorrectly predicts a positive class, related to Type I error) and a low rate of false negatives (errors where the model fails to predict a positive class, related to Type II error). However, it's important to note that while precision and recall give insights into the model's performance, they do not directly measure Type I and Type II errors.

Overall, the proposed finetued KNN model arises itself as a robust model with a high degree of generalisation making suitable for the classification purposes of the present report.
